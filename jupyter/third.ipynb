Mutable and immutable  objects

```python
my_tup = (1, "bflmpsvz", ["a", "b", "c"], {"alpha", "beta", "gamma"})

print("try to modify element[0] - number")
try:
    my_tup[0] = 2
except Exception as e1:    
    print(str(e1))

print()
print("try to modify element[1] - part of a string ")
try:
    my_tup[1][2] = "L"
except Exception as e2:
    print(str(e2))
    
print()
print("try to modify element[2] - part of a list ")
my_tup[2][1] = "B"

print()
print("try to modify element[3] - set")
my_tup[3].add("delta")

print()
print(my_tup)
```
